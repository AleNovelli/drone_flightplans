```python
import pandas as pd
import pymap3d as pm
import numpy as np
import json
```

```python
filename = "2025_12_campaign"
df = pd.read_csv(filename+".csv")
df["Name"] = df["Name"].str.lower().str.replace(r"launchpad", "landing_site", regex=True)
telescopes={}
landing_site={}
site={}
```

```python
for tel_name in ["satp1", "satp2", "satp3", "class1", "class2"]:
    telescope_data = df[df["Name"].str.lower().str.startswith(tel_name)]
    lat=telescope_data["Latitude"]
    lon=telescope_data["Longitude"]
    alt=telescope_data["Ellipsoidal height"]
    
    x,y,z = pm.geodetic2ecef(lat=lat, lon=lon, alt=alt)
    
    X=np.mean(x)
    Y=np.mean(y)
    Z=np.mean(z)

    LAT, LON, ALT = pm.ecef2geodetic(X, Y, Z)

    telescopes[tel_name]= {
        "lat":LAT,
        "lon":LON,
        "alt":ALT
    }


telescope_data = df[df["Name"].str.lower().str.startswith("landing_site")]
lat=telescope_data["Latitude"]
lon=telescope_data["Longitude"]
alt=telescope_data["Ellipsoidal height"]

x,y,z = pm.geodetic2ecef(lat=lat, lon=lon, alt=alt)

X=np.mean(x)
Y=np.mean(y)
Z=np.mean(z)

LAT, LON, ALT = pm.ecef2geodetic(X, Y, Z)

landing_site["landing_site"]= {
    "lat":LAT,
    "lon":LON,
    "alt":ALT
}

site["telescopes"] = telescopes
site["landing_site"] = landing_site

with open(filename+".json", "w") as f:
    json.dump(site, f, indent=4) 
```
